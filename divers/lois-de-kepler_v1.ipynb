```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

################################################################################
#Fonctions de calcul d'aires
################################################################################

def calcul_aire(J, Jour):
    """Cette fonction calcule l'aire en faisant le calcul de la surface d'un triangle
    Soleil-Position J-PositionJ+1  l'aide du produit vectoriel SP(j)^SP(j+1)"""
    aire = 0
    for j in range (-J, J) :
        x, y, z = X[Jour+j], Y[Jour+j], Z[Jour+j]
        x1, y1, z1 = X[Jour+j+1], Y[Jour+j+1], Z[Jour+j+1]
        surface_triangle = 0.5*np.sqrt((y*z1-z*y1)**2+ (z*x1-x*z1)**2+(x*y1-y*x1)**2)
        aire = aire + surface_triangle
    return(aire)

################################################################################
#Fonction pour tracer la surface parcoururue
################################################################################

def trace_parcourue(J, Jour, couleur):
    for j in range (-J,J+1) :
        ax.plot([0,X[Jour+j]],[0,Y[Jour+j]],color=couleur, linewidth = 3)
    return()

################################################################################
#LECTURE DU FICHIER
################################################################################
Terre = 'https://capytale2.ac-paris.fr/web/sites/default/files/2020-12-13-19-08-44/ac-versailles-1/mln_laurent.grosbois4/terre.csv'
Jupiter = 'https://capytale2.ac-paris.fr/web/sites/default/files/2020-12-13-19-08-44/ac-versailles-1/mln_laurent.grosbois4/jupiter.csv'
Mars = 'https://capytale2.ac-paris.fr/web/sites/default/files/2020-12-13-19-08-44/ac-versailles-1/mln_laurent.grosbois4/mars.csv'
Mercure = 'https://capytale2.ac-paris.fr/web/sites/default/files/2020-12-13-19-09-20/ac-versailles-1/mln_laurent.grosbois4/mercure.csv'
Neptune = 'https://capytale2.ac-paris.fr/web/sites/default/files/2020-12-13-19-09-20/ac-versailles-1/mln_laurent.grosbois4/neptune.csv'
Saturne = 'https://capytale2.ac-paris.fr/web/sites/default/files/2020-12-13-19-09-20/ac-versailles-1/mln_laurent.grosbois4/saturne.csv'
Uranus ='https://capytale2.ac-paris.fr/web/sites/default/files/2020-12-13-19-09-20/ac-versailles-1/mln_laurent.grosbois4/uranus.csv'
Venus ='https://capytale2.ac-paris.fr/web/sites/default/files/2020-12-13-19-09-20/ac-versailles-1/mln_laurent.grosbois4/venus.csv'


fname = input("Saisir le nom du fichier sans l'extension (.csv)")
#Lecture du fichier
Data = pd.read_csv(fname+'.csv', sep=',', header=0)
#Modification de la ligne date du format objet vers le format datetime
Data[' Date'] = pd.to_datetime(Data[' Date'])
# Rcupration de pas de temps (en jours) = temps entre 2 enregistrements conscutifs
DeltaT = (Data[' Date'][1]-Data[' Date'][0]).days

#Stocker les variables dans une matrice
# Position
X = Data[' X (au)']
Y = Data[' Y (au)']
Z = Data[' Z (au)']
M = np.sqrt(X**2+Y**2+Z**2)

# Vitesse
Vx = Data[' Xp (au/day)']
Vy = Data[' Yp (au/day)']
Vz = Data[' Zp (au/day)']
V = np.sqrt(Vx**2+Vy**2+Vz**2)

#Calcul des distances  l'aphlie et au prihlie
Perihelie = np.min(M[5:-5])
Aphelie = np.max(M[5:-5])

#Recuprer les index des points  l'aphlie et au prihlie
ID_Perihelie = np.where(M==Perihelie)[0][0]
ID_Aphelie = np.where(M==Aphelie)[0][0]

###############################################################################
#Trac de la trajectoire
###############################################################################
fig=plt.figure(figsize=(16,10))

ax = fig.add_subplot(121)
#Tracer l'ellipse et placer les lgendes
ax.plot(X,Y, color='k')
ax.text(X[ID_Aphelie]-0.01,Y[ID_Aphelie]-0.05,"A",fontsize = 16,color='b')
ax.text(X[ID_Perihelie]-0.05,Y[ID_Perihelie]+0.005,"P",fontsize = 16,color='r')
ax.scatter(0,0,color='k', s=100)
ax.text(0.03,-0.03,'S', fontsize = 16)

# Tracer les diffrentes zones balayes
J=4
trace_parcourue(J,ID_Perihelie,'m')
trace_parcourue(J,ID_Aphelie,'g')

#Paramtres figure
ax.axis('equal')
plt.grid(True)


#Lgendes
plt.title("Trac de l'orbite de " + str(fname))
ax.set_ylabel(r'$y$ (ua)')
ax.set_xlabel('$x$ (ua)')


################################################################################
#Troisime loi de Kepler
################################################################################

#Calcul du demi-grand axe
a=((M[ID_Aphelie]+M[ID_Perihelie])/2)

#Dtermination de la priode de rvolution en jours
j=20 # pour s'tre suffisamment loign du point de dpart...
marge=max(M)/100
while abs(X[j]-X[0])>marge or abs(Y[j]-Y[0])>marge or abs(Z[j]-Z[0])>marge:
    j = j+1
T=j*DeltaT

ax1=fig.add_subplot(122,frameon=False)
plt.xlim(0,13)
plt.ylim(0,11)
plt.xticks([])
plt.yticks([])
ax1.text(0.5,10.5,"Au point P : ", color='m', fontsize=14)
ax1.text(1.,10," - Distance  l'astre attracteur : "+str(round(M[ID_Perihelie],5))+ " $\mathrm{ua}$")
ax1.text(1.,9.5," - Valeur de la vitesse : "+str(round(V[ID_Perihelie],5))+ " $\mathrm{ua\cdot j^{-1}}$")
ax1.text(1.,9," - Aire balaye par le segment [SP] entre "+ str(-J*DeltaT) +" jours et +" + str(+J*DeltaT) + " jours autour de P : "+str(round(calcul_aire(J,ID_Perihelie),7))+ " $\mathrm{ua^2}$")

ax1.text(0.5,7.5,"Au point A : ", color='g', fontsize=14)
ax1.text(1.,7," - Distance  l'astre attracteur : "+str(round(M[ID_Aphelie],5))+ " $\mathrm{ua}$")
ax1.text(1.,6.5," - Valeur de la vitesse : "+str(round(V[ID_Aphelie],5))+ " $\mathrm{ua\cdot j^{-1}}$")
ax1.text(1.,6," - Aire balaye par le segment [SA] entre "+ str(-J*DeltaT) +" jours et +" + str(+J*DeltaT) + " jours autour de A : "+str(round(calcul_aire(J,ID_Aphelie),7))+ " $\mathrm{ua^2}$")

ax1.text(0.5,4.5,"Troisime loi de Kepler ", color='r', fontsize=14)
ax1.text(1,4,"Demi-grand axe $a$ : "+'$a$=%.2e'%(a)+" ua")
ax1.text(1,3.5,"Priode de rvolution, T = "+str(T)+' jours')

plt.savefig('Kepler_'+(fname)+'.png')

plt.show()



```
