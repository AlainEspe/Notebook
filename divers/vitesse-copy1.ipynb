```python
import matplotlib.pyplot as plt
import numpy as np

#nom du fichier à traiter
fichier=input("Quel est le nom du fichier de pointage (sans l'extension .txt)?") +".txt"

t,x,t,y = np.genfromtxt (fichier, skip_header =1, unpack = True)

m=len(t)-1

for i in range(0,m,3) :
    vx = (x[i+1]-x[i])/(t[i+1]-t[i])
    vy = (y[i+1]-y[i])/(t[i+1]-t[i])
    plt.arrow(x[i],y[i],0.1*vx,0.1*vy,
    head_width=0.06,width=0.015,
    color = "green")


plt.xlim(xmin =0 ,xmax=np.max(x)+0.5)
plt.ylim(ymin =0 ,ymax=np.max(y)+0.2)
plt.grid()
plt.plot(x,y,'ro-', label = "y=f(x)")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.legend()
plt.title("Trajectoire du ballon")
plt.show()
```
