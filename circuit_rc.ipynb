```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')

c = 1000 * 10**(-6)
E = 6
r = 1000

t = np.linspace(0,10,1000)
Uc = E*(1- np.exp((-1/(r*c))*t))



plt.plot(t,Uc,label='Charge du condensateur')

print('Tau',1/(r*c))

plt.xlabel('temps (s)')
plt.title('circuit RC')
plt.legend()
plt.show()

t = np.linspace(0,10,100)
Uc = E*np.exp((-1/(r*c))*t)

plt.plot(t,Uc,label='décharge du condensateur')

print('Tau',1/(r*c))

plt.xlabel('temps (s)')
plt.title('circuit RC')
plt.legend()
plt.show()
```
